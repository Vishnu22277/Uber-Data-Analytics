## **About**  
**ncr_ride_bookings.csv**

This dataset captures Uber ride booking activity in the National Capital Region (NCR), offering a snapshot of urban mobility patterns. It’s ideal for exploratory data analysis (EDA), trend detection, and visual storytelling.
## **Likely Columns** (based on typical ride data)
- booking_id: Unique identifier for each ride
- timestamp: Date and time of booking
- pickup_location: Area or zone where the ride was requested
- dropoff_location: Destination zone
- ride_type: Type of Uber service (e.g., UberGo, UberX)
- booking_status: Status of the ride (e.g., Completed, Cancelled, No-show)
- fare_amount: Estimated or actual fare
- payment_method: Mode of payment (e.g., Cash, Card, Wallet)
## **Use Cases**
- Analyze peak booking hours and days
- Compare ride types and their popularity
- Identify high-demand pickup zones
- Visualize booking status distribution
- Detect patterns in cancellations or no-shows
## **Sample Questions You Can Explore**
- What time of day sees the most ride bookings?
- Which locations have the highest ride demand?
- How does booking status vary by ride type or time?
- Are there seasonal or weekly trends in ride activity?

 ## **EDA**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

# **Ride Status Distribution**


```python
plt.figure(figsize=(8, 4))
sns.countplot(x='booking_status', data=df, hue='booking_status', palette='Set2', legend=False)
plt.title("Booking Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#**Vehicle Type Usage**

```python
vehicle_counts = df['vehicle_type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(vehicle_counts.values, labels=vehicle_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Vehicle Type Distribution")
plt.axis('equal')
plt.show()
```

#Pickup vs Drop Location Heatmap


```python
route_matrix = df.pivot_table(index='pickup_location',
                              columns='drop_location',
                              values='booking_id',
                              aggfunc='count',
                              fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(route_matrix, cmap='YlGnBu', linewidths=0.5)
plt.title("Ride Frequency: Pickup vs Drop Location")
plt.xlabel("Drop Location")
plt.ylabel("Pickup Location")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

# **Ride Frequency by Hour**

```python
df['hour'] = pd.to_datetime(df['time'], format='%H:%M:%S', errors='coerce').dt.hour

plt.figure(figsize=(10, 4))
sns.countplot(x='hour', data=df, hue='hour', palette='coolwarm', legend=False)
plt.title("Ride Frequency by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Rides")
plt.tight_layout()
plt.show()
```

# **Booking Value Distribution**

```python
df['booking_value'] = pd.to_numeric(df['booking_value'], errors='coerce')

plt.figure(figsize=(8, 4))
sns.histplot(df['booking_value'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Distribution of Booking Value")
plt.xlabel("Booking Value (₹)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
